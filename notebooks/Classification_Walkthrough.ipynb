#Setup and Data Import

```python
def disp(df, n=5):
  print(df.shape)
  display(df.head(n))
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv("spambase.csv", header=0)
disp(df)
```

```python
df.describe()
```

#First Run (Using Naive Bayes)

```python
import sklearn
from sklearn.naive_bayes import GaussianNB

X = df.iloc[:, :-1]
y = df.iloc[:, -1]

nb_model = GaussianNB()
nb_model.fit(X, y)
y_pred = nb_model.predict(X)

preds = pd.DataFrame({"y_actual": y, "y_pred": y_pred})

display(preds.head(5))
display(preds.tail(5))
```

# Metrics

```python
from sklearn import metrics

print(f"Accuracy: {metrics.accuracy_score(y, y_pred)}")
print(f"Precision: {metrics.precision_score(y, y_pred)}")
print(f"Recall: {metrics.recall_score(y, y_pred)}")
print(f"F1-Score: {metrics.f1_score(y, y_pred)}")
```

```python
print(metrics.classification_report(y, y_pred))
```

# Train/Test Split

```python
# Split data into train and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=314)
```

```python
# Fit NB model to data, make predictions, and calculate accuracy metrics
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
y_pred = nb_model.predict(X_test)
report = metrics.classification_report(y_test, y_pred)

print(report)
```

# Hyperparameters (Using K-Nearest Neighbors)

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=10)
knn_model.fit(X_train, y_train)
y_pred = knn_model.predict(X_test)
report = metrics.classification_report(y_test, y_pred)

print(report)
```

# Hyperparameter Tuning

```python
# Basic Hyperparameter Tuning with For Loop

scores = []

for k in range(1, 30):
  knn_model = KNeighborsClassifier(n_neighbors=k)
  knn_model.fit(X_train, y_train)
  y_pred = knn_model.predict(X_test)
  prec = metrics.precision_score(y_test, y_pred)
  rec = metrics.recall_score(y_test, y_pred)
  f1 = metrics.f1_score(y_test, y_pred)
  scores.append([k, prec, rec, f1])

scores_df = pd.DataFrame(scores, columns=["k", "precision", "recall", "f1"])
  
print(scores_df)
```

# K-Fold Cross Validation

```python
from sklearn.model_selection import KFold, cross_val_score

cv = KFold(n_splits=10, random_state=314, shuffle=True)
knn_model = KNeighborsClassifier(n_neighbors=1)
f1_scores = cross_val_score(knn_model, X, y, scoring="f1", cv=cv, n_jobs=-1)

print(np.mean(f1_scores))
```

# Grid Search with CV

```python
# Put it together! (K-Fold CV within a Grid Search)

from sklearn.model_selection import GridSearchCV


def run_grid_search(X, y, model, params, scoring="f1", n_folds=10):
  cv = GridSearchCV(model, params, scoring=scoring, cv=n_folds)
  cv.fit(X, y)
  results = cv.cv_results_
  cols = ["param_" + name for name in params.keys()] + ["mean_test_score", "std_test_score", "rank_test_score"]
  results_df = pd.DataFrame(results)[cols]
  display(results_df.sort_values(["rank_test_score"]))

  return results_df
```

```python
# Use grid search function with custom inputs

knn_model = KNeighborsClassifier()
params = {"n_neighbors": [1, 2, 3, 4, 5]}

results_df = run_grid_search(X, y, knn_model, params, scoring="f1", n_folds=10)
```

# Create Functions

```python
def preprocess_data(data_path):
  df = pd.read_csv(data_path, header=0)
  df = df.rename(columns={"char_freq_[": "char_freq_sqbracket"})
  X = df.iloc[:, :-1]
  y = df.iloc[:, -1]
  ss = StandardScaler()
  ss.fit(X)
  X_ss = pd.DataFrame(ss.transform(X))
  X_ss.columns = X.columns

  return X_ss, y


def run_grid_search(X, y, model, params, scoring="f1", n_folds=10):
  cv = GridSearchCV(model, params, scoring=scoring, cv=n_folds)
  cv.fit(X, y)
  best_model = cv.best_estimator_
  best_params = cv.best_params_
  results = cv.cv_results_
  param_cols = ["param_" + name for name in params.keys()]
  score_cols = ["mean_test_score", "std_test_score", "rank_test_score"]
  results_df = pd.DataFrame(results)[param_cols + score_cols]\
                .sort_values(["rank_test_score"])
  display(results_df.head())

  return best_model, best_params
```

```python
# Try them out
X, y = preprocess_data(data_path="spambase.csv")

knn_model = KNeighborsClassifier()
params = {"n_neighbors": range(1, 5)}
knn_best, knn_params = run_grid_search(X, y, knn_model, params, scoring="f1", n_folds=10)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(solver="liblinear")
params = {"penalty": ["l1", "l2"]}
lr_best, lr_params = run_grid_search(X, y, lr_model, params, scoring="f1", n_folds=10)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier()
params = {"max_depth": [1, 5, 10], "min_samples_split": [2, 5, 10], "min_samples_leaf": [2, 5, 10]}
dt_best, dt_params = run_grid_search(X, y, dt_model, params, scoring="f1", n_folds=10)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
params = {"max_depth": [1, 5, 10], "min_samples_split": [2, 5, 10], "min_samples_leaf": [2, 5, 10]}
rf_best, rf_params = run_grid_search(X, y, rf_model, params, scoring="f1", n_folds=10)
```

# XGBoost

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(n_estimators=100, learning_rate=0.1, subsample=0.9)
params = {
    "max_depth": [1, 5, 10],
    "colsample_bytree": [0.5, 1.0],
    "reg_alpha": [0.1, 1.0]
}
xgb_best, xgb_params = run_grid_search(X, y, xgb_model, params, scoring="f1", n_folds=10)
```

# Support Vector Machines

```python
from sklearn.svm import SVC

svc_model = SVC(probability=True, random_state=314)
params = {"kernel": ["linear", "poly", "rbf", "sigmoid"]}
svc_best, svc_params = run_grid_search(X, y, svc_model, params, scoring="f1", n_folds=10)
```

# Model Ensembles

```python
from sklearn.ensemble import VotingClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=314)

estimators = [
              ("knn", knn_best), 
              ("lr", lr_best), 
              ("dt", dt_best), 
              ("rf", rf_best), 
              ("xgb", xgb_best), 
              ("svc", svc_best)
]
ensemble = VotingClassifier(estimators, voting="soft")
ensemble.fit(X_train, y_train)
y_pred = ensemble.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
print(metrics.f1_score(y_test, y_pred))
```

# Model Serialization

```python
import joblib

joblib.dump(xgb_best, "xgb_model.pkl")
```

```python
def validate_model(model_path, data_path):
  model = joblib.load(model_path)
  X_val, y_val = preprocess_data(data_path=data_path)
  y_pred = model.predict(X_val)
  print(metrics.classification_report(y_val, y_pred))
  print(f"Accuracy: {round(100*metrics.accuracy_score(y_val, y_pred), 1)}%")
  print(f"F1: {round(100*metrics.f1_score(y_val, y_pred), 1)}%")
```

```python
model_path = "xgb_model.pkl"
data_path = "spambase_val.csv"
validate_model(model_path, data_path)
```
